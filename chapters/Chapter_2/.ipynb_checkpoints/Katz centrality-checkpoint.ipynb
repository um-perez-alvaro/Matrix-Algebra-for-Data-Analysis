# KATZ CENTRALITY MEASURE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

## 1. TOY EXAMPLE <a class="anchor" id="toy"></a>

Let's start with a toy example. We'll build the graph from its adjacency matrix.

```python
A = np.zeros((13,13))
A[0,[1]] = 1
A[1,[2,5,6]] = 1
A[2,[1,7,8,3]] = 1
A[3,[9,10,4]] = 1
A[4,[3,11,12]] = 1
A[6,1] = 1
A[7,2] = 1
A[8,10] = 1
A[9,7] = 1
A[10,3] = 1
A[12,10] = 1
A
```

```python
G = nx.from_numpy_array(A, create_using=nx.DiGraph())
G
```

To plot the graph, let's assign positions to the nodes.

```python
# node position dictionary
pos = {
    0 : [0,0],
    1 : [2,0],
    2 : [4,0],
    3 : [6,0],
    4 : [8,0],
    5 : [2,2],
    6 : [2,-2],
    7 : [4,2],
    8 : [4,-2],
    9 : [6,2],
    10 : [6,-2],
    11 : [8,2],
    12 : [8,-2]
}
```

Now we can plot the graph.

```python
plt.figure(figsize=(12, 4))
nx.draw(G, 
        pos, 
        with_labels=True, 
        node_size=700, 
        node_color='skyblue', 
        width = 2,
        font_size=10, 
        font_weight='bold', 
        edge_color='black')
```

Now let's compute the Katz centralities of the nodes. Recall that the Katz centrality vector is given by

$$
    c_{\rm Katz} = e^T\left((I_n-\alpha A)^{-1}-I_n\right),
$$

```python
# Katz centralities
alpha = 0.1
n = len(A)
e = np.ones((1,n))
cKatz = e.dot(np.linalg.inv(np.eye(n)-alpha*A)-np.eye(n))
cKatz = cKatz.flatten()
cKatz
```

Finally, just for fun, let's plot the graph and color the nodes by their centrality values.

```python
plt.figure(figsize=(12, 4))

# Normalize centralities for color mapping
norm = plt.Normalize(vmin=min(cKatz), vmax=max(cKatz))

# Create a color map
cmap = plt.cm.viridis

# Get node colors from centralities
node_colors = cmap(norm(cKatz))

# Draw the graph
nx.draw(G, 
        pos, 
        with_labels=True, 
        node_size=700, 
        node_color=node_colors, 
        width = 2,
        font_size=10, 
        font_weight='bold', 
        edge_color='black')

# Add colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, label='Katz Centrality')
```

## 2. US AIRPORT TRANSPORTATION NETWORK <a class="anchor" id="airports"></a>

The next example uses the US air transportation network. Each node represents an airport, and there is an edge between nodes if there is a direct flight between the airports. The following DataFrame contains all the edges of the graph

```python
# load the data from the Github repo
path = 'https://raw.githubusercontent.com/um-perez-alvaro/Matrix-Algebra-for-Data-Analysis/main/datasets/data/chapter%202/flights.csv'
flights = pd.read_csv(path)
flights
```

We can create the graph from this DataFrame and then extract the adjacency matrix.

```python
# Create the graph from the edges DataFrame
G = nx.from_pandas_edgelist(flights, 'ORIGIN', 'DEST')
A = nx.adjacency_matrix(G).todense()
A
```

Using the adjacency matrix, we can compute the Katz centralities.

```python
alpha = 0.01
n = len(A)
e = np.ones((1,n))
cKatz = e.dot(np.linalg.inv(np.eye(n)-alpha*A)-np.eye(n))
cKatz = cKatz.flatten()
```

We can store the centralities in a DataFrame.

```python
centralities_df = pd.DataFrame(index=G.nodes,data=cKatz,columns=['centrality'])
centralities_df
```

The following DataFrame contains information about each airport, including its name and location.

```python
# load the data from the Github repo
path = 'https://raw.githubusercontent.com/um-perez-alvaro/Matrix-Algebra-for-Data-Analysis/main/datasets/data/chapter%202/airports.csv'
airports = pd.read_csv(path,index_col='CODE')
airports
```

We can add the centralities to this DataFrame.

```python
airports['centrality'] = centralities_df
```

sort the airports by their Katz centrality values, and see which are the top 10 airports

```python
airports.sort_values(ascending=False,by='centrality').head(10)
```

Finally, let's plot the graph and color the nodes according to their centrality.

```python
plt.figure(figsize=(12, 4))

# Normalize centralities for color mapping
norm = plt.Normalize(vmin=min(cKatz), vmax=max(cKatz))

# Create a color map
cmap = plt.cm.viridis

# Get node colors from centralities
node_colors = cmap(norm(cKatz))

# node positions 
pos = airports[['LONGITUDE', 'LATITUDE']].T.to_dict('list')

# Draw the graph
nx.draw(G, 
        pos, 
        #with_labels=True, 
        node_size=50, 
        node_color=node_colors, 
        width = 0.01,
        font_size=10, 
        font_weight='bold', 
        edge_color='black')

# Add colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, label='Katz Centrality')

plt.xlim([-130,-66])
plt.ylim([23,50])
```
